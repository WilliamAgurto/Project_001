Algunos campos, como belongs_to_collection, production_companies y otros (ver diccionario de datos) están anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila, ¡deberán desanidarlos para poder y unirlos al dataset de nuevo hacer alguna de las consultas de la API! O bien buscar la manera de acceder a esos datos sin desanidarlos.

```python
#according to data in excel there are five columns wich has a list of dicctionaries. So, I take them out in order to unnest them.
# the columns are 'belongs_to_collection','production_companies','genres','spoken_languages' and 'production_countries'
dictionaries = df_movies[['belongs_to_collection','production_companies','genres','spoken_languages','production_countries']]
dictionaries.head()
```

```python
def pull_out_data(df, columna):
    # Function to extract values in string, list or dictionary
    def pull_out_value(valor):
        if isinstance(valor, str):
            # If it is string, pull out values with "regular expression (re)"
            pattern = r"'name': '([^']*)'"
            coincidencias = re.findall(pattern, valor)
            return coincidencias if coincidencias else np.nan
        elif isinstance(valor, list):
            # if it is a list, pull out values from each value to the list
            valores = []
            for elemento in valor:
                if 'name' in elemento:
                    valores.append(elemento['name'])
            return valores if valores else np.nan
        elif isinstance(valor, dict):
            # If it is a dictionary, pull out the values from 'name' key 
            if 'name' in valor:
                return valor['name'] if valor['name'] else np.nan
        return np.nan
    
    # applying function on the DataFrame
    df[columna] = df[columna].apply(pull_out_value)
    # registers are list-type. So i order to have a string of them i 'll apply 'applymap' function.
    df = df.applymap(lambda x: ', '.join(x) if isinstance(x, list) else x)
    return df
```

```python
#checking 'dictionaries' out
dictionaries.head()
```

```python
# Applying the function to pull put data from 'belongs_to_collection','production_companies','genres','spoken_languages' and 'production_countries' columns
dictionaries = pull_out_data(dictionaries, 'belongs_to_collection')
dictionaries = pull_out_data(dictionaries, 'production_companies')
dictionaries = pull_out_data(dictionaries, 'genres')
dictionaries = pull_out_data(dictionaries, 'spoken_languages')
dictionaries = pull_out_data(dictionaries, 'production_countries')
#movie results
dictionaries
```

```python
#Empty values will be filled by nan value
dictionaries.spoken_languages[dictionaries.spoken_languages==''] = np.nan
dictionaries
```

```python
# Turning values columns into new ones 
df_movies.drop(columns=['belongs_to_collection','production_companies','genres','spoken_languages','production_countries'], inplace=True)
df_movies =pd.concat([df_movies,dictionaries], axis=1)
df_movies.head()
```

2. Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0

```python
print(f'budget nulls: {df_movies.budget.isnull().sum()}')
print(f'revenue nulls: {df_movies.revenue.isnull().sum()}')
#There is no null in budget column so i'll appy only for revenue column
df_movies.revenue.fillna(0, inplace=True)
print(f'new revenue nulls : {df_movies.revenue.isnull().sum()}')
```

3.Los valores nulos del campo release date deben eliminarse.

```python
print(f'there are {df_movies.release_date.isnull().sum()} nulls on release_date column ')
#droping nulls values
df_movies.dropna(subset=['release_date'], inplace=True)
print(f'Now there are {df_movies.release_date.isnull().sum()} nulls on release_date column ')
```

4. De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

```python
# removing blanck space on strip method
df_movies['release_date'] = df_movies['release_date'].str.strip()
# Turning release_date into date format on pd.to_datetime() function
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], format='%Y-%m-%d')
```

**According to the mistake showed, the are a register which has a value '1'. But there re some others, too such as '12' ans '22'.**

```python
print(np.array(df_movies[df_movies.release_date == str(22)]))
print(np.array(df_movies[df_movies.release_date == str(12)]))
print(np.array(df_movies[df_movies.release_date == str(1)]))
```

**As we can see. those values were not bee correctly upload because date is there but in another column. Because of the fact that they're just three, I'll drop them.**

```python
df_movies = df_movies.drop(df_movies[df_movies.release_date == '1'].index)
df_movies = df_movies.drop(df_movies[df_movies.release_date == '12'].index)
df_movies = df_movies.drop(df_movies[df_movies.release_date == '22'].index)
```

```python
# And Now I can Turning release_date into date format on pd.to_datetime() function
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], format='%Y-%m-%d')
```

5. Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python
#turn budget column into float
df_movies.budget = df_movies.budget.astype(float)
# creating and computing return column
df_movies['return'] = np.where(df_movies['budget'] == 0, 0, df_movies['revenue'] / df_movies['budget'])
df_movies.head()
```
